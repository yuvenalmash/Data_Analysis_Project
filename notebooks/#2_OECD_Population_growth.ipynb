# OECD countries' "Population" growth over the last decade

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import openpyxl
```

```python
# import data set
file = '../data/WEOOct2020alla.xlsx'
df = pd.read_excel(file, engine='openpyxl')
```

```python
# Filter the DataFrame to include only rows where the "country group name" is "Advanced economies"
df_oecd = df[df['Country Group Name'] == 'Advanced economies'].copy()
```

```python
# Select the data for the advanced economies
gdp_data = df[(df["Country Group Name"] == "Advanced economies") & (df["WEO Subject Code"] == "NGDPD")]
gdppc_data = df[(df["Country Group Name"] == "Advanced economies") & (df["WEO Subject Code"] == "PPPGDP")]



```

```python
gdp_data_values = gdp_data.values[0][-13:-3]
gdppc_data_values = gdppc_data.values[0][-13:-3]
years = gdp_data.columns[-13:-3]
```

```python
# Calculate the population estimates
population_data = gdp_data_values / gdppc_data_values
```

```python
# Plot the population growth
plt.figure(figsize=(10, 6))
plt.plot(years, population_data)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth in OECD Countries')
plt.grid(True)
plt.show()
```
